## Human Development Index (HDI)
The Human Development indicator contains three pamrameter, life expectancy, eduction and icome. The HDI has been computed following the same formula as used by the UNDP, namely the geometric mean of the three dimensions: health (life expectancy), education (geometric average of actual and expected educational attainment measured by years of schooling) and income (gross national income (GNI) per capita). [Source: https://www.doi.org/10.1553/populationyearbook2015s241]

For all of the following considerations, data from the following source was used:

## Human Development Index in the EU under five scenarios
SSP1 Sustainability Scenario: describes the world in which sustainable developments are a priority and environmental limits are respected, national and international cooperation is strengthened.  
SSP2 Middle of the road: global development largely follows historical trends, with some countries making great progress while others lag behind.  
SSP3 Regional rivalry: The scenario is dominated by nationalism and regional conflicts, with domestic politics and regional issues taking precedence. International cooperation has low priority, including in the area of climate protection.  
SSP4 Inequality: The gap between the layers of society is growing, but the social differences between countries are also growing.  
SSP5 Fossil-fueled development: The world is driven by the economic success of both industrialized and emerging countries, but growth goes hand in hand with the exploitation of fossil resources and an energy-intensive lifestyle.

```python
import pyam as pym

hdi_data = pym.IamDataFrame(data="SSP-Extensions_Human_Development_Index_v1.0.xlsx")

hdi_eu = hdi_data.filter(region="European Union (R9)",variable="Human Development Index [weighted by Population]")
eu_plt = hdi_eu.plot(color="scenario",title=" HDI weigthed by Population of the European Union",ylabel='HDI' )
```

The plot shows the HDI for the European Union weighted by population up to year 2075. The largest short-term increases in HDI occur in the scenarios SSP1 and SSP5. In the long term, however, the highest HDI is achieved with scenario SSP2. Scenarios SSP3 and SSP4 produce the smallest increase in HDI. This means that in the long term it is best for the EU to focus together on sustainable solutions. Nationalism and the support of social inequality inhibit the growth of the HDI.

## Scenarios that reach the limit of 0.85 by 2070
This chapter examines which of the 5 scenarios exceeds the HDI limit of 0.85. This can be used to determine which scenarios make sense for a specific goal or specific analyses. The HDI values ‚Äã‚Äãof the European Union are considered here, weighted by population. In this example, the categorize method is used to determine whether a scenario reaches the limit.

```python
import pyam as pym

hdi_data = pym.IamDataFrame(data="SSP-Extensions_Human_Development_Index_v1.0.xlsx")

hdi_eu = hdi_data.filter(region="European Union (R9)",variable="Human Development Index [weighted by Population]")

hdi_eu.set_meta(meta="above 0.85", name = "HDI-Category")
hdi_eu.categorize("HDI-Category","below 0.85",criteria = {"Human Development Index [weighted by Population]":{"up": 0.85,"year":2070}})
print(hdi_eu)
hdi_eu.plot(color="scenario",linestyle="HDI-Category",title=" HDI weigthed by Population of the European Union",ylabel='HDI' )
```

```python
hdi_eu.meta
```

In this example you can see that the scenarios SSP3 and SSP4 never reach the limit of 0.85. The scenarios SSP1, SSP2 and SSP5 reach the selected limit by the year 2070. Output 3 shows the HDI-Category that were calculated.

## Differences in HDI for different weights using scenario SSP2
This example examines the differences between the individual weighting models. For the EU there are three weightings for the HDI: unweighted mean, weighted by GDP, weighted by Population. In this example for the European Union, only the SSP2 scenario is considered. It can be seen from the plot that there is a difference between the individual weightings. This must be taken into account when working with the data and when individual regions are to be compared with each other. In this example, the subtract method is used to calculate the difference.  DI

```python
import pyam as pym
import pandas as pd
hdi_data = pym.IamDataFrame(data="SSP-Extensions_Human_Development_Index_v1.0.xlsx")

hdi_eu = hdi_data.filter(region="European Union (R9)",scenario="SSP2")


hdi_eu.subtract("Human Development Index [unweighted mean]","Human Development Index [weighted by GDP]",
                "Difference: (unweighted mean) - (weighted by GDP)",append=True)
hdi_eu.subtract("Human Development Index [unweighted mean]","Human Development Index [weighted by Population]",
                "Difference: (unweighted mean) - (weighted by Population)",append=True)

hdi_eu_diff_GDP = hdi_eu.filter(variable="Difference: (unweighted mean) - (weighted by GDP)")
hdi_eu_diff_Pop = hdi_eu.filter(variable="Difference: (unweighted mean) - (weighted by Population)")
frames = [hdi_eu_diff_GDP,hdi_eu_diff_Pop]
hdi_polt = pym.concat(frames)
hdi_polt.plot(color="variable",title=" HDI weigthed by Population of the European Union",legend={"loc": "outside right"} )
```
